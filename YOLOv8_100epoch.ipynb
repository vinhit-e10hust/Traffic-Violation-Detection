```python
from ultralytics import YOLO
import matplotlib.pyplot as plt

SEED = 42
model = YOLO("yolov8m.pt")

results = model.train(
    data=r"C:/Users/Acer/Downloads/Project CV/extracted_data/data.yaml",
    epochs=100,
    imgsz=640,
    batch=16,
    device=0,
    project="runs",
    name="traffic_100_epochs",
    exist_ok=True,
    seed=SEED,
    patience=10,
    close_mosaic=10,
)

# Print tiến trình huấn luyện
print("Training completed. Results:")
print(results)

# Quan sát output sau khi huấn luyện
print("Best mAP@0.5:", results.best_map50)
print("Best mAP@0.5:0.95:", results.best_map50_95)

# Trực quan theo timeseries
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(results.epoch, results.box_loss, label="Box Loss")
plt.plot(results.epoch, results.cls_loss, label="Class Loss")
plt.plot(results.epoch, results.obj_loss, label="Object Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Loss Curves")
plt.legend()
plt.subplot(1, 2, 2)
plt.plot(results.epoch, results.map50, label="mAP@0.5")
plt.plot(results.epoch, results.map50_95, label="mAP@0.5:0.95")
plt.xlabel("Epoch")
plt.ylabel("mAP")
plt.title("mAP Curves")
plt.legend()
plt.tight_layout()
plt.show()

# Trực quan hóa timeseries của Precion và Recall
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(results.epoch, results.precision, label="Precision")
plt.xlabel("Epoch")
plt.ylabel("Precision")
plt.title("Precision Curve")
plt.legend()
plt.subplot(1, 2, 2)
plt.plot(results.epoch, results.recall, label="Recall")
plt.xlabel("Epoch")
plt.ylabel("Recall")
plt.title("Recall Curve")
plt.legend()
plt.tight_layout()
plt.show()


# Demo dự đoán ngẫu nhiên trên một vài ảnh test
test_images = [
    r"C:/Users/Acer/Downloads/Project CV/extracted_data/images/test/6722.jpg",
    r"C:/Users/Acer/Downloads/Project CV/extracted_data/images/test/6684.jpg",
    r"C:/Users/Acer/Downloads/Project CV/extracted_data/images/test/6638.jpg",
]
for img_path in test_images:
    result = model.predict(img_path, conf=0.25)
    result.show()


```
